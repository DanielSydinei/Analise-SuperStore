```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"data\Sample - Superstore.csv", encoding='latin1')
df.head()
```

```python
df.info()
```

```python
df.groupby('Category')['Sales'].sum().sort_values(ascending=True).plot(kind='bar')
plt.title('Categoria de Produto Mais Vendida')
plt.xlabel('Categorias', color='blue')
plt.ylabel('Vendas Totais', color='green')
plt.xticks(rotation=0)
```

```python
df.groupby('Region')['Profit'].sum().sort_values(ascending=True).plot(kind='barh')  
plt.title('Lucro por Região')
plt.xlabel('Lucro', color='green')    
plt.ylabel('Região', color='blue')
plt.xticks(rotation=0)
```

```python
df.groupby('City')['Profit'].sum().nlargest(5).sort_values(ascending=True).plot(kind='bar')
plt.title('Lucro por Cidade')
plt.xlabel('Cidade', color='blue')    
plt.ylabel('Lucro Total', color='green')
plt.xticks(rotation=0)
```

```python
df["Segment"].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Clientes que mais compraram')
```
